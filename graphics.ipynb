```python
from math import sqrt
import os
import pylab
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import scipy as sp

start_path = '/home/asd/PycharmProjects/search_in_graph/results'
d = {6: 2.5706, 11: 2.2281, 16: 2.1314,  21: 2.0860, 26: 2.0555,
     31: 2.0423, 36: 2.0301, 41: 2.0211, 51: 2.0086, 101: 1.9840}
```

```python
def readfile(name, pos):
    y = []
    with open(os.path.join(start_path, name)) as f:
        for line in f.readlines():
            y.append(float(line.split()[pos]))
    return sorted(y)
```

```python
x = [[30 + i for i in range(0, 1001, 50)], [30 + i for i in range(0, 3001, 50)]]
y = [readfile('result-dijkstra-random', 3), readfile('result-ford-random', 3), readfile('result-levit-random', 3)]
s = [readfile('result-dijkstra-random', 4), readfile('result-ford-random', 4), readfile('result-levit-random', 4)]
n = [readfile('result-dijkstra-random', 5), readfile('result-ford-random', 5), readfile('result-levit-random', 5)]

fig, ax = plt.subplots(figsize=(18, 13))

ax.set_title('Рандомный граф с n вершинами и n * 10 ребрами', fontsize=20)
ax.set_xlabel("Количество вершин", fontsize=15)        
ax.set_ylabel("Время", fontsize=15)
ax.scatter(x[1], y[0], color="orange", label="dijkstra", linewidth=0.01)    
ax.scatter(x[1], y[1], color="red", label="ford", linewidth=0.01)     
ax.scatter(x[1], y[2], color="gold", label="levit", linewidth=0.01)
for i in range(len(x[1])):
    ax.bar(x[1][i], y[0][i], yerr=d[n[0][i]] * s[0][i] / sqrt(n[0][i]), ecolor="black", alpha=0, linewidth=2)
    ax.bar(x[1][i], y[1][i], yerr=d[n[1][i]] * s[1][i] / sqrt(n[1][i]), ecolor="black", alpha=0, linewidth=2)
    ax.bar(x[1][i], y[2][i], yerr=d[n[2][i]] * s[2][i] / sqrt(n[2][i]), ecolor="black", alpha=0, linewidth=2)
ax.grid(which="major", linewidth=1)
ax.grid(which="minor", linestyle="--", color="gray", linewidth=0.5)

ax.legend(loc=2, prop={'size': 20})
plt.show()
```
